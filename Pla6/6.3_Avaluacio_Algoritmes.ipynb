# 6.3 Avaluació d'algoritmes

Sigui quin sigui l'algoritme que apliquem per classificar o predir resultats, sempre necessitarem tenir algun tipus d'avaluació dels resultats, saber com es comportarà, com de bé ho farà; sobretot si volem comparar dos o més algoritmes per decidir quin és el millor per un tipus concret de dades.

Què vol dir avaluar resultats? Doncs saber quan precís és un algoritme assolint els seus objectius o, en altres paraules, quin percentatge d'encerts i errors té. No hi ha prou amb implementar un algoritme de classificació i llençar-lo sobre un grup de dades, hem de saber si ens podrem fiar dels resultats que obtingui. 

Hi ha diverses eines per mesurar els resultats dels algoritmes de machine learning. En la majoria de casos  en l'aprenentatge supervisat volem saber si hem classificat bé els resultats, si el sistema aprèn suficientment bé de les dades d'aprenentatge com per obtenir una precisió suficientment elevada. El llindar que decidim per considerar que una precisió és suficientment bona dependrà de les dades i la complexitat del problema. Per exemple, si només entrarem dades de 2 classes diferents, i llencéssim una moneda per decidir el resultat de cada mostra d'entrada, l'estadística ens diu que obtindríem una precisió del 50%; sembla evident, doncs, que qualsevol algoritme que superi aquest valor es podrà considerar que funciona bé perquè obtindrà millors resultats que si ho féssim aleatòriament. Ara bé, en tindríem prou amb una precisió del 55%? o del 60%? Voldria dir que 5-6 de cada 10 mostres estarien ben classificades però que, per altra banda, que 4-5 de cada 10 no ho estarien. Ens podríem fiar d'uns resultats així?  Ens serien útils? Probablement no, quan més s'apropi la precisió al 100% millor que millor, tot i que ja us avancem que aquest resultat s'aconsegueix en poques ocasions dins el món del machine learning.

## Mètriques d'avaluació
Abans de passar a veure mètodes i mètriques d'avaluació, hi ha un parell de conceptes bàsics que hem de tenir molt clars: què vol dir encert i què vol dir error. A priori diríem que encert és tot allò que ha estat correctament classificat, mentre que error seria la resta. Fins aquí, fàcil. Però dins dels errors, són tots iguals? No, i és important saber-ne la diferència, sobretot per poder millorar tant l'algoritme com les dades d'aprenentatge que estiguem utilitzant.

Si tenim un sistema amb dues categories diferents (A i B), quan avaluem els resultats possibles respecte a la categoria A, podrem trobar 4 tipus diferents de resultats:

- Positius: Encerts. Mostres de la classe A que han estat classificades per l'algoritme com classe A.
- Negatius: Encerts. Mostres de la classe B que no han estat classificades per l'algoritme com classe A
- Falsos Positius: Errors. Mostres de la classe B que han estat classificades per l'algoritme com classe A.
- Falsos Negatius: Errors. Mostres de la classe A que no han estat classificades per l'algoritme com classe A.

## Validació creuada

Un cop entès que hem de buscar els valors òptims de les mètriques d'avaluació, quines dades hem de fer servir per calcular-les? 

En tot aprenentatge supervisat tindrem un grup de mostres d'entrenament, és a dir, un grup de mostres que hauran estat etiquetades correctament. Normalment es divideixen aquestes mostres en dos grups: un grup per entrenar el sistema i un altre grup per testar (o validar) els resultats (així sabrem si la sortida de l'algoritme és la correcta o no). 

Però hem de tenir en compte que poques vegades les dades son pures i perfectes, en la majoria dels casos tenim soroll, fluctuacions o errors en les dades que ens poden afectar els resultats. Què passaria si just el grup d'entrenament o el de validació tingués un nombre elevat d'aquestes impureses en les dades? Doncs que els resultats de les mètriques d'avaluació no serien gaire ajustats a la realitat. Per exemple, imagineu què passaria si en un sistema amb 4 possibles categories escollim un grup d'entrenament amb només mostres d'una de les 4 categories.

Per evitar aquestes possibles desviacions existeixen les tècniques de **cross-validation**, que en català ho traduïm com tècniques de *validació creuada*.

L'objectiu principal de la validació creuada és garantir que els resultats que obtinguem siguin independents de la partició de dades d'entrenament i validació (o test). S'utilitza molt per validar models generats en projectes d'aprenentatge supervisat. Bàsicament, consisteix en repetir i calcular la mitjana aritmètica de les mètriques d'avaluació sobre diferents particions de les dades.

Una de les tècniques de cross-validation més populars és la coneguda com **K-Fold**, la qual, bàsicament, divideix tot el conjunt de mostres en K grups, i fa K iteracions de càlculs. En la primera iteració el sistema agafa el primer grup com grup de test i la resta de grups com entrenament (K-1 grups), entrena el sistema i calcula la mètrica d'avaluació; en la segona iteració repeteix els passos de la primera però fent servir el segon grup com grup de test i la resta com grups d'entrenament... Al final tindrem K valors de mètriques d'avaluació dels quals calcularem la mitjana per donar el resultat final.

Existeixen diverses variacions del K-Folds: validació creuada aleatòria (enlloc d'escollir mostres consecutives per testar a cada iteració, s'escullen mostres aleatòries), validació creuada deixant-ne un fora (es fa servir només una mostra per validar en cada iteració), etc. Podeu trobar més informació a les referències del final d'aquesta secció. 


```python
# EXEMPLE PRÀCTIC: AVALUANT I COMPARANT ALGORITMES
# COMPARATIVA ENTRE DOS ALGORITMES DE CLASSIFICACIO SOBRE UN MATEIX DATASET
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import svm

# carreguem dades
data = datasets.load_breast_cancer()
# dividim entre conjunt d'aprenentatge i de test
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=0)

print("Mostres totals {}, aprenentatge {}, test {}".format(data.data.shape[0], X_train.shape[0], X_test.shape[0]))

def create_and_test_classifier(model, x_train, y_train, x_test, y_test):
    # creem un classificador
    model.fit(x_train, y_train)
    # executem una validació creuada amb les dades d'aprenentatge
    validacio_creuada_accuracy = cross_val_score(model, x_train, y_train, cv=5)
    print('validacio creuada 5-fold: {}, mitjana {}'.format(validacio_creuada_accuracy, sum(validacio_creuada_accuracy) / 5))

    # predim les dades de test
    y_predict = model.predict(x_test)
    
    print('Matriu de confusio\n', confusion_matrix(y_test, y_predict))

    return accuracy_score(y_test, y_predict)

            
# mostrem el rendiment de l'algoritme amb diferents quantitats de  dades
models = [DecisionTreeClassifier(random_state=0), svm.SVC(kernel='linear', C=1), svm.SVC(kernel='rbf')]
model_names = ['Decision tree', 'SVM linear', 'SVM RBF']
for model_name, current_model in zip(model_names, models):
    
    print("\nModel {}------------------".format(model_name))
    
    accuracy = create_and_test_classifier(current_model, X_train, y_train, X_test, y_test)
    
    print("Exactitud (accuracy) {}".format(accuracy))
    
    
```
