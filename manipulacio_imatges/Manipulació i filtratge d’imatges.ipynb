## Manipulació i filtratge d'imatges##

Les imatges digitals es representen com un conjunt de nombres, on cada nombre representa el valor d'un píxel

Les imatges en blanc i negre, tenen un valor de color per píxel que sol anar del 0 al 255 (8 bits de color). Les imatges color, tenen 3 valors per píxel (el canal verd, vermell i blau), que també ocupen 8 bits cadascun (3 bytes per píxel)

OpenCV treballa amb els canals en ordre BGR (Blau Verd i Vermell)

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('lena.jpg')

# Si mostrem la imatge tal i com está al disc, no es veurà correctament (el canal blau i verd estan canviats)
plt.subplot(1,2,1)
plt.title('BGR')
plt.imshow(image)

# si invertim l'ordre dels canals, ja es veurà correctament la imatge
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.subplot(1,2,2)
plt.title('RGB')
plt.imshow(image_rgb)
```

## Filtres bàsics d'intensitat ##

Donada una imatge amb nivells d'intensitat, podem seleccionar els valors més brillants comparant cada píxel amb un valor donat

```python
image = cv2.imread('lena.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_bn = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.subplot(2,2,1); plt.title('RGB'); plt.imshow(image_rgb)
plt.subplot(2,2,2); plt.title('BN'); plt.imshow(image_bn, 'gray')

# comparem cada pixel de la imatge amb un valor fix
image_thresholded = image_bn > 127
# el mateix, però fent servir la funció de les OpenCV
ret,image_thresholded_opencv = cv2.threshold(image_bn,127,255,cv2.THRESH_BINARY)

plt.subplot(2,2,3); plt.title('Thresholded'); plt.imshow(image_thresholded, 'gray')
plt.subplot(2,2,4); plt.title('Thresholded Opencv'); plt.imshow(image_thresholded_opencv, 'gray')
```
