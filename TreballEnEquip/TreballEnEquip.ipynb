# Treball en equip
Tal i com us vam avançar en la presentació del curs, algunes competències transversals d'aquest curs inclouen el treball en equip, i algunes competències específiques la implantació de la metodologia Scrum a través de eines com Trello i Git.

Aquest capítol recull una introducció a tots aquests conceptes.

## SCRUM
Scrum és un mètode Agile per desenvolupar projectes, normalment projectes de software.

Originalment, els projectes de software feien servir una metodologia anomenada _en cascada_ on el projecte tenia diverses fases que havien de ser completades totalment abans de poder passar a la següent.

Tot i que la metodologia funcionava, convertia qualsevol projecte en un procés llarg i molts farregós per a possibles canvis de requeriments o gestió d'errors en els mateixos. A més, la majoria de vegades cada fase del procés el portaven equips diferents, el que feia que quan el l'equip de codificació trobava algun requerients sense sentit, o bé alguna manera de millorar el producte final, fos quasi impossible aplicar un canvi, el procés havia de continuar i, en tot cas, si es feia una segona iteració es tindrien en compte els canvis.

Amb el temps i, sobretot, amb l'aparició de cada cop més software a desenvolupar i aplicar en quasi tots els àmbits de la nostra vida, aquest tipus de metodologia va anar fracassant. Es posava tant d'esforç en la planifiació i documentació del cada cicle de desenvolupament, que es perdia el sentit del que realment importava pels clients finals i quines eren les millors solucions per a cada projecte. Al 2001, un grup de desenvolupadors insatisfets amb el paradigma de desenvolupament de software del moment va escriure el manifest Agile, que va revolucionar i canviar la manera de treballar des de llavors.

En la metodologia Agile, enlloc de tenir totes els requeriments detallats al màxim i tancats des de l'inici del projecte, s'aplica un un procés incremental i interactiu, on es deixa la responsabilitat compartida entre tot l'equip. La idea principal és aprofitar l'expertesa de cada membre de l'equip deixant-los decidir com trobar i aplicar la millor solució a cada problema i requeriment, a la vegada que es fa responsable a cadascun d'ells de la seva parcel.la de treball, d'aquesta manera tothom se sent més involucrat en el projecte, a la vegada que es treu el màxim profit de la sabiesa de tot l'equip. I també es canvia totalment la planificació del projecte, amb una metodologia Agile es canvien els temps llargs de mesos o anys per obtenir una versió final del producte, amb temps curts de poques setmanes on cada vegada es treu una versió més refinada i commpleta del producte final.

Hi ha diverses metodologies àgils: Scrum, Kanban, XP (eXtreme Programming)... Scrum és pot ser una de les més populars perquè emfatitza les versions de software pràctiques (que funcionen des de la primera versió), la col·laboració dins l'equip i entre equips, i com la flexibilitat per adaptar-se a les realitats emergents de negocis.

En la seva metodologia el projecte es subdivideix en petits objectius (tasques del backlog) a assolir en un curt periode de temps (sprint) que típicament és de dues setmanes. Les tasques es poden agrupar per objectiu comú (epic) per facilitar la revisió del progrés de tot el projecte.

Resumint molt la metodologia, podríem dir que al principi de cada Sprint l'equip decideix en quines tasques treballarà, durant l'Sprint diàriament l'equip fa una reunió curta de màxim 15 minuts (Standup meeting) on s'informa de com van les tasques, i a cada final de Sprint les tasques acabades s'entreguen i es crea una nova versió del producte.  

Si voleu saber més detalls de la metodologia Scrum podeu trobar diferents cursos a les referències del final del capítol.

Hi ha diferents rols involucrats en l'equip de Scrum, però a efectes pràctics per a aquest curs farem una versió molt i molt resumida de la metodologia basant-nos només amb alguns principis de la mateixa i fent servir eines com el Trello i el GIT.

Tasques a realitzar en cada Pla/Projecte del curs:
1. Decidir objectius i tasques que necessita el projecte
2. Escriure-les en el backlog del Trello (columna Backlog) com noves targetes (una per tasca diferent)
3. A l'inici de l'Sprint decidir quines tasques es desenvoluparan i qui les farà (assignar la persona i la data final a cada tasca/targeta)
4. Durant l'sprint cada persona treballa en les tasques que té assignades
5. Diàriament (o bé cada 2-3 dies en aquest curs) l'equip es reuneix i contesta a 3 preguntes: 
 - en què vas treballar ahir? (o des de l'última reunió)
 - en què treballaràs avui? (o fins la propera reunió)
 - en què necessites ajuda?
6. Quan una tasca es comença es treu del Backlog i es mou a la columna de Doing/Fent-se
7. Quan una tasca s'acaba es mou a la columna de Done/Feta (en la vida real tindríem també columnes de  verificació, testeig o QA abans de donar-la per acabada)




## Trello

Trello és una eina visual per organitzar tasques que, a més a més, permet treballar de manera col·laborativa entre diferentes persones i equips.

Té versions per ordinador i també Apps per mòbils i tauletes.

És una eina molt intuïtiva on bàsicament hi ha taulers, llistes i targetes.

  - A: Un tauler (_board_) representa un projecte.
  - B: Les llistes (_list_) són columnes per organitzar i veure l'estat de les tasques. Es poden crear tantes com vulgueu, però una configuració típica per a qualsevol projecte és tenir-ne 3 que representen els 3 estats possibles d'una tasca: pendent de fer (_To Do_ o _Backlog_), fent-se (_Doing_), i fet (_Done_).
  - C: Les targetes (_card_) són les tasques a realitzar i permeten afegir tanta informació com vulgueu.


Us deixem l'organització interna de cada equip a les vostres mans, i aquí us donem alguna recomanació per organitzar-vos:
1. Primer creeu un tauler (_board_) pel vostre projecte. El tauler us permet crear també un equip de treball o bé convidar a col·laboradors al tauler. 
2. Entreu dins del tauler i configureu les columnes que fareu servir: Backlog, Fent-se, Fet
3. Per començar el projecte primer heu de crear les tasques a realitzar: 
 1. Una targeta nova per a cada tasca.
 2. Si una tasca és molt llarga, podeu dividir-la en diverses subtasques o bé afegir una llista de subtasques dins la mateixa targeta.
 3. Assigneu cada tasca a la persona que correspongui.
 4. Assigneu data de finalització a la tasca.
 5. Assigneu una etiqueta a la tasca per a una millor visualització dels objectius (per exemple, una etiqueta per projecte o pla).
4. Quan comenceu una tasca moveu-la a la columna Fent-se.
5. Quan acabeu una tasca moveu-la a la columna Fet.

En qualsevol moment podeu afegir comentaris a una targeta per comunicar-vos amb els companys d'equip o per deixar constància de temes relacionats amb la tasca.


## Git

Git és un sistema lliure, de codi obert i distribuit pel control de versions de programació. Serveix per controlar els canvis en el codi durant el seu procés de desenvolupament, a la vegada que està dissenyat per coordinar les tasques dins un mateix equip de programació. Serveix per tot tipus de tamany de projectes, petits i grans.

Al principi de cada projecte o pla, us haureu de baixar la versió corresponent del Git.
Al final de cada projecte haureu de tenir la versió final pujada al GitLab per poder avaluar-la.
