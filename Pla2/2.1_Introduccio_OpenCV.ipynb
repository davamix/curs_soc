# 2.1 Introducció a l'OpenCV: matriu per imatges

## Visio per Computador
La Visió per Computador (VC d'ara en endavant), és la branca de la intel·ligència artificial que s'encarrega del processament i anàlisi d'imatges, amb l'objectiu d'extreure informació de les mateixes i poder oferir solucions a problemes del món real.

El **processament d'imatges** s'encarrega de tractar les imatges i processar-les per extreure'n algun tipus d'informació. Per exemple, podem passar una imatge en color a blanc i negre per poder analitzar-la més fàcilment, podem enfosquir una imatge per remarcar aquells objectes més il.luminats, etc.
L'**anànàlisi d'imatges** utilitza la informació continguda en imatges processades per resoldre problemes del món real com, per exemple, detectar objectes concrets en una imatge.

VC s'aplica actualment a molts camps presents en el nostre dia a dia, des del control de trànsit, la seguretat en els edificis, creació i anàlisi de proves mèdiques, la robòtica, etc.

Alguns exemples reals poden ser el reconeixement de cares que veiem en Facebook, la detecció i reconeixement d'emprentes digitals per desbloquejar un mòbil, detectar somriures en temps real per fer una fotografia en una càmara digital,  extreure models 3D d'objectes, detecció d'unitats defectuoses en una cadena de muntatge, comptatge de cèl·lules sanguínies en una analítica de sang, etc. La llista podria ser llarguíssima, i cada dia s'incorporen noves solucions basades en VC.


## Sobre OpenCV
Dins la VC i machine learning, OpenCV (_Open Source Computer Vision_) és una de les llibreries de software més utilitzades arreu del món tant a nivell educatiu com a nivell privat. Una de les raons principals és que proporciona un entorn de codi obert, per al desenvolupament, fàcil d'utilitzar i a la vegada molt eficient pel que fa a càlcul computacional. I també s'ha de tenir en compte que la llibreria és molt robusta (han tret només 4 versions en els seus 20 anys de la seva història) i que inclou més de 2500 algoritmes optimitzats de VC i machine learning; des dels clàssics fins als més nous i innovadors.

Aquí llistem algunes de les seves característiques més rellevants:
- Llibreria de software de codi obert.
- Orientada principalment a aplicacions de visió i processament de imatges en temps real.
- Utilitza operacions vectorials per a la majoria de les seves funcions, el que fa de l’OpenCV una llibreria molt eficient en termes de computació.
- Multiplataforma: suporta Linux, MacOS, Windows, iOS i Android.
- Disponible per diferents llenguatges de programació: Tot i que el motor està desenvolupat en C i C++, permet desenvolupar en C, C++, Python i Java, entre d'altres llenguatges.
- Lllicència de producte BSD: OpenCV va ser creada per proporcionar i facilitar una infraestructura comuna per a les aplicacions de VC i per accelerar l'ús d'aquesta en productes comercials, per la qual cosa ofereix una llicència pels negocis fàcil d'utilitzar, a la vegada que permet modificar el codi per adaptar-lo a cada objectiu comercial.
- La llibreria s'utilitza àmpliament en empreses privades, grups de recerca i organismes governamentals: OpenCV té més de 47.000 persones en la seva comunitat d'usuaris i s'estima que el nombre de baixades supera els 18 milions.

Què necessitem primer de tot? Imatges, imatges digitals que hem de carregar d'alguna manera a l'ordinador per poder començar a treballar amb elles.


```python
# EXEMPLE PRÀCTIC: CARREGANT IMATGES

import cv2 # Càrrega de la llibreria OpenCV
from matplotlib import pyplot as plt # Càrrega de la llibreria de visualització integrada en Jupyter

# Lectura de la imatge de disc, forçant que sigui en blanc i negre
imatge = cv2.imread('Imatges/lena.jpg', cv2.IMREAD_GRAYSCALE)

# Mostrem la imatge, indicant que volem fer servir el mapa de color de grisos
plt.imshow(imatge, cmap='gray')
plt.title('Lena')
```

Un cop tenim la imatge carregada en el sistema, anem a veure com podem manipular-la amb algun exemple bàsic

```python
# EXEMPLE PRÀCTIC: MANIPULACIÓ BÀSICA D'IMATGES
import cv2
from matplotlib import pyplot as plt

# Lectura de la imatge de disc, forçant que sigui en blanc i negre
imatge = cv2.imread('Imatges/lena.jpg', cv2.IMREAD_GRAYSCALE)

# Com que les imatges es representen amb enters, podem modificar-les operant amb elles com si fòssin 
# nombres, per exemple, sumant una constant. 
imatge_aclarida = imatge + 20

# Anem a mostrar les dues imatges utilitzant la funció "plt.subplot" que permet fer un dibuix en 2 eixos.
# Indiquem amb el primer paràmetre el nombre de files, i amb el segon el nombre de columnes.
# El tercer paràmetre indica la posició (o cel·la) dins el dibuix de 2 eixos on volem mostrar la imatge.
plt.subplot(1,2,1)
plt.title('Lena') # Mostrem la imatge original
plt.imshow(imatge, cmap='gray')

# Mostrem la imatge aclarida
plt.subplot(1,2,2)
plt.title('Lena aclarida')
plt.imshow(imatge_aclarida, cmap='gray')

# Mostrem la figura més gran per veure les imatges
fig = plt.gcf()
fig.set_size_inches(12.5, 12.5)

# un cop verifiquem que el resultat és correcte, desem el resultat a disc
cv2.imwrite('Imatges/imatge_aclarida.jpg', imatge_aclarida)

```

Ja hem fet els nostres primers passos, sabem com carregar imatges, sabem que es poden manipular o canviar, desar i visualitzar. Fins i tot hem vist que les imatges es representen amb enters, i que podem fer operacions aritmètiques amb aquests enters, però com és possible? Com es representa una imatge internament? Us ho avancem: amb *matrius numèriques*!

## Matrius per imatges
Les imatges digitals es representen computacionalment com matrius numèriques, on cada píxel és un element de la matriu, i el seu valor indica el color del mateix.

### Imatges binàries
Per començar amb un exemple molt bàsic, pensem amb una imatge binària o booleana, on cada píxel només pot tenir valor 0 o 1. Un element amb valor 0 indica que el píxel és de color negre, el 1 indica color blanc, Utilitzant aquesta representació, la següent imatge del gat Fèlix (a l'esquerra) tindria la seva representació matricial de 35x35 que veiem a la dreta.



### Imatges en escala de grisos
Les imatges amb escales de grisos, segueixen la mateixa idea que les binàries però utilitzant valors dins del rang [0-255] per representar els diferents nivells de gris (0 continua sent negre, i el 255 és el blanc). Això dóna un total de 256 valors diferents. Veiem-ne un exemple.


```python
#EXEMPLE PRÀCTIC: REPRESENTACIÓ MATRICIAL D'UNA IMATGE AMB ESCALA DE GRISOS

imatge_matricial = cv2.imread('Imatges/lena.jpg', cv2.IMREAD_GRAYSCALE)
plt.subplot(1,2,1)
plt.title('Lena') # Mostrem la imatge original
plt.imshow(imatge_matricial, cmap='gray')

# Podem veure que les imatges es representen amb una matriu "imprimint" directament la imatge
print('Pixels de la IMATGE ORIGINAL:')
print(imatge_matricial)

# També podem veure les seves dimensions
print("Mida imatge {}".format(imatge_matricial.shape))

# I, fins i tot, podem recòrrer tots els píxels de la imatge amb una iteració aniuada, 
# recorrent files i columnes per aplicar qualsevol operació que necessitem o per fer alguna cerca
for i in range(imatge_matricial.shape[0]):
    for j in range(imatge_matricial.shape[1]):
        # accedim a cada píxel i incrementem en 20 el seu valor
        # això seria equivalent al que hem fet en l'exercici anterior
        imatge_matricial[i,j] += 20
        
plt.subplot(1,2,2)
plt.title("Lena aclarida")
plt.imshow(imatge_matricial, cmap='gray')

print('\nPíxels de la IMATGE ACLARIDA:')
print(imatge_matricial)
print("Mida imatge {}".format(imatge_matricial.shape))
```

Tot i que no és fàcil llegir una matriu de 512x512 valors, sí que podem fixar-nos en els valors de les cantonades. Les submatrius 3x3 que es mostren quan imprimim la matriu de la imatge. Veieu com els valors de la matriu varien després d'aplicar la iteració aniuada per incrementar els valors? I també és interessant fixar-nos en els valors numèrics i comparar-los amb la imatge corresponent per veure com mostren els diferents nivells de grisos.

Aquest tipus de representació amb escala de grisos acostuma a ser suficient per imatges web, però tingueu en compte que hi ha algunes imatges en escala de grisos que necessiten encara més resolució o detall, és a dir que necessiten més de 256 valors diferents per definir i detectar al màxim les característiques diferents. Les imatges mèdiques serien un d'aquests casos especials.

### Imatges en color
Però anem un pas més endavant, si ja fem servir 256 valors per una imatge amb escala de grisos, com representem una imatge en color? Doncs amb **3 matrius**; cada matriu especificarà el nivell de Vermell, Verd i Blau que conté cada píxel de la imatge. I cada element pot tenir 256 possibles valors dins dels rang 0-255, per indicar el nivell de cadascun dels 3 colors. Aquest sistema és conegut com **sistema RGB**, per les seves inicials en anglès dels colors vermell (Red), verd (Green) i blau (Blue), i pot arribar a representar fins a 16777216 colors diferents.

El sistema RGB és el sistema més popular per representar colors en ordinadors, però hem de tenir en compte que l'OpenCV fa servir el sistema BGR, el qual ve a ser el mateix, però canviant l'ordre dels colors. El per què d'aquest canvi no està clar, molta gent creu que és per una raó històrica, perquè quan es va començar a desenvolupar el codi de l'OpenCV el sistema de color en format BGR era el més popular entre els fabricants de càmeres i els proveïdors de software, per la qual cosa els desenvolupadors van escollir aquest sistema i així ha quedat fins a dia d'avui. Sigui quina sigui la raó de fer servir el sistema BGR, no té gaire importància si treballem només amb OpenCV i Python, però sí que ho haurem de tenir en compte si fem servir software extern per visualitzar alguna imatge manipulada i/o desada amb OpenCV perquè els canals R(vermell) i B(blau) estaran intercanviats (el sistema extern esperarà una imatge amb sistema RGB mentre que nosaltres li donarem una amb sistema BGR).




```python
#EXEMPLE PRÀCTIC: VISUALITZANT SISTEMA RGB/BGR 

import cv2
from matplotlib import pyplot as plt

# Per defecte, OpenCV carrega les imatges assumint que estan en ordre BGR. 
# Tanmateix, la majoria de vegades les imatges estan en ordre RGB
imatge = cv2.imread('Imatges/lena.jpg')

# Si mostrem la imatge tal i com está al disc, no es veurà correctament (el canal blau i verd estan canviats)
plt.figure()
plt.subplot(1,2,1)
plt.title('BGR')
plt.imshow(imatge)

# Si invertim l'ordre dels canals, ja es veurà correctament la imatge
image_rgb = cv2.cvtColor(imatge, cv2.COLOR_BGR2RGB)
plt.subplot(1,2,2)
plt.title('RGB')
plt.imshow(image_rgb)

fig = plt.gcf()
fig.set_size_inches(8, 12)

# També podem manipular cadascun dels canals de la imatge color
canal_blau = imatge[:,:, 0]
canal_verd = imatge[:,:, 1]
canal_vermell = imatge[:,:, 2]

plt.figure()
plt.subplot(1,3,1)
plt.title('Blau')
plt.imshow(canal_blau, cmap='gray')

plt.subplot(1,3,2)
plt.title('Verd')
plt.imshow(canal_verd, cmap='gray')

plt.subplot(1,3,3)
plt.title('Vermell')
plt.imshow(canal_vermell, cmap='gray')
# Mostrem la figura més gran per veure les imatges
fig = plt.gcf()
fig.set_size_inches(12, 12)


# Tingeu en compte que la imatge original té un to vermellós, per això el canal vermell 
# és més intens (blanc) que la resta
```

Ara que ja sabem com es desen i interpreten les imatge en color, anem a veure com passar una imatge en color a escala de grisos. Aquest, tot i que sembli estrany a primera vista, és el primer pas que haurem de fer en la majoria de processaments d'imatge. Convertir una imatge en color a blanc i negre (o a escala de grisos) ens permetrà analitzar molt millor la imatge resultant en la majoria dels casos.



```python
#EXEMPLE PRÀCTIC: CONVERTIR IMATGE EN COLOR A ESCALA DE GRISOS

import cv2
from matplotlib import pyplot as plt

imatge = cv2.imread('Imatges/lena.jpg')

imatge_grisos = cv2.cvtColor(imatge, cv2.COLOR_BGR2GRAY)

plt.title('Grisos')
plt.imshow(imatge_grisos, cmap='gray')

```

### Transformacions bàsiques de matrius
Les matrius ens permeten fer molts tipus d'operacions, i aquestes operacions es veuran reflectides en la imatge que representen. Per exemple, què passaria si apliquem la matriu transposta a la matriu original de la imatge? 


```python
#EXEMPLE PRÀCTIC: MATRIU TRANSPOSTA

imatge_transposada = imatge_grisos.transpose()
plt.subplot(1,2,1)
plt.title('Grisos')
plt.imshow(imatge_grisos, cmap='gray')

plt.subplot(1,2,2)
plt.title('Transposada')
plt.imshow(imatge_transposada, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(8, 8)


# Per a veure diferents transformacions geomètriques, consultar:
#https://docs.opencv.org/trunk/da/d6e/tutorial_py_geometric_transformations.html
```
